## 재귀호출

```python

count = 10

def openBox():
    global count
    print('종이상자를 엽니다. ^^')
    count -= 1
    if count <= 0:
        print('반지는 넣고 반환합니다.')
        return
    openBox() #종이상자를 엽니다
    print('종이 상자를 닫습니다')
```

```python
openBox()
```

# 재귀호출로 덧셈

```python
def addNumber(num):
    if num <= 1:
        return 1
    else:
        return num + addNumber(num-1)
    
print(addNumber(3))
```

```python
factValue = 1
for n in range(10, 0 ,-1):
    factValue *= n

print('10*9*8..*2*1', factValue)
```

# 팩토리얼 재귀호출

```python
def factorial(num):
    if num <= 1:
        return 1
    else:
        return num * factorial(num -1)
print('10! = ' , factorial(1200))
```

### 우주선 발사

```python
import time
def countdown(n):
    if n == 0:
        print('우주선 발사')
    else:
        print(n)
        #time.sleep(1)
        countdown( n - 1)
countdown(10)
```

```python
def printStar(n):
    if n> 0:
        printStar(n-1)
        print('*' * n)

printStar(5)
```

```python
def gugudan(dan, num):
    print(f'{dan}x{num}={dan*num:2d}')
    if num < 9 :
        gugudan(dan, num+1)

for dan in range(2,10):
    print(f'{dan}단 시작 ------')
    gugudan(dan,1)
```

### N 제곱

```python
tab = ' '
def pow(x,n):
    global tab
    tab += ' '
    if n == 0:
        return 1
    print(tab, f'{x}x{x}^{n}-{1}')
    return x * pow(x, n-1)
print('2^10')
print('답 -->',pow(2,10))
```

### 피보나치

```python
def fibonacci(n):
    if n == 0: return 0
    elif n == 1: return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print('피보나치 수 --> 0 1', end=' ')
for i in range(2,41):
    print(fibonaci(i),end= ' ')
```

```python
def palindrome(pStr): # kayak
    if len(pStr) <= 1:
        return True
    
    if pStr[0] != pStr[-1]: # k == k a==a
        return False

    return palindrome(pStr[1:len(pStr)-1])
```

```python
strAry = ['reaver', 'kayak', 'borrow or job','주유소의 소유주', '이번주 주번이 너야' ,'살금살금' ]
```

```python
for testStr in strAry:
    print(testStr, end='--> ')
    testStr = testStr.lower().replace(' ', ' ')# 영어 대문자 x , #공백x
    if palindrome(testStr) == True:
        print('회문입니다')
    else:
        print('회문 아니오 ㅡㅡ ;')
```
